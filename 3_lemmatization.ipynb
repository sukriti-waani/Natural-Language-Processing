## **`Lemmatization`**
**Lemmatization** is a Natural Language Processing (NLP) technique that converts a word into its **base or dictionary form (called lemma)** using **vocabulary and grammatical analysis**.

The output of lemmatization is usually a **meaningful word found in the dictionary**.

### Example

| Word | Lemmatized Word |
|-----|----------------|
| running | run |
| studies | study |
| better | good |
| cars | car |

## Comparison: Stemming vs Lemmatization

| Feature | Stemming | Lemmatization |
|-------|-------|-------|
| Method | Removes suffixes using rules | Uses dictionary and grammar |
| Output | May not be a real word | Produces meaningful word |
| Speed | Faster | Slower |
| Accuracy | Less accurate | More accurate |

---

## Example Comparison

| Word | Stemming | Lemmatization |
|------|---------|--------------|
| studies | studi | study |
| running | run | run |
| better | better | good |
| flying | fli | fly |

---

#### In Short

- **Stemming** removes suffixes using simple rules.
- **Lemmatization** uses dictionary and grammar to return meaningful base words.
- Lemmatization is **more accurate but slower** than stemming.

### **`WordNet Lemmatizer`**

**WordNet Lemmatizer** is a lemmatization tool provided by **NLTK (Natural Language Toolkit)** that reduces a word to its **base or dictionary form (lemma)** using the **WordNet lexical database**.

WordNet is a large English dictionary that groups words into **sets of synonyms (called synsets)** and stores relationships between words. The WordNet Lemmatizer uses this database to return **meaningful root words based on the word's part of speech (POS)**.

### Example

| Word | Lemma |
|-----|------|
| running | run |
| studies | study |
| better | good |
| cars | car |

### Key Points

- Uses **WordNet dictionary**
- Produces **meaningful base words**
- Considers **Part of Speech (POS)**
- More **accurate than stemming**


### Where WordNet Lemmatizer is Used

- Text preprocessing in **Machine Learning**
- **Search engines** for better keyword matching
- **Chatbots and virtual assistants**
- **Sentiment analysis**
- **Information retrieval systems**

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
lemmatizer.lemmatize("going")
```

```python
'''
POS- Noun-n
verb-v
adjective-a
adverb-r
'''
lemmatizer.lemmatize("going", pos='v')
```

```python
words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]
```

```python
for word in words:
    print(word + " -----> " + lemmatizer.lemmatize(word, pos='n'))
```

```python
lemmatizer.lemmatize("goes", pos='v')
```
