**NLTK (Natural Language Toolkit)** is a Python library used for Natural Language Processing (NLP). It provides tools and datasets to help computers process, analyze, and understand human language (text).

In short:
NLTK = A Python toolkit for working with and analyzing text data.

### **Tokenization**
Tokenization is the process of breaking a sentence or text into smaller parts called tokens (words, characters, or sentences).

Example:

Sentence:

"I love learning NLP"

After tokenization:

["I", "love", "learning", "NLP"]

So each word becomes a token.

Simple idea:
👉 Tokenization = splitting text into words or pieces.

### **Corpus**
A corpus is a large collection of text documents used for training or analyzing language models.

Example:

A dataset containing many sentences:

"I love NLP"
"Machine learning is interesting"
"AI is the future"

All these texts together form a corpus.

Simple idea:
👉 Corpus = a big collection of text data.

### **Vocabulary**
Vocabulary is the set of all unique words present in a corpus.

Example:

Corpus:

"I love NLP"
"I love AI"

Vocabulary:

["I", "love", "NLP", "AI"]

Here duplicates are removed.

Simple idea:
👉 Vocabulary = all unique words in the dataset.

Quick Summary

Term	--->>>   Meaning
- Tokenization	--->>>   Splitting text into words
- Corpus	--->>>   Collection of many text documents
- Vocabulary	--->>>   Unique words present in the corpus

```python
corpus = """Hello Welcome, to Sukriti's learning.
I am learning NLP.
"""
```

```python
print(corpus)
```

```python
import nltk
nltk.download('punkt')   #  #Punkt is a pre-trained model in NLTK used for sentence tokenization by identifying punctuation marks that indicate sentence boundaries.
```

```python
# Tokenization
# sentence --> paragraphs


# Importing the sent_tokenize function from the nltk.tokenize module

# nltk -> Natural Language Toolkit (Python library used for NLP tasks)

# tokenize -> A submodule in NLTK that contains functions
#             used to split text into smaller units like
#             sentences or words.

# sent_tokenize -> A function used to break a paragraph
#                  or large text into individual sentences.

# Example:
# Text: "I love NLP. It is interesting."
# After sent_tokenize:
# ["I love NLP.", "It is interesting."]

from nltk.tokenize import sent_tokenize
```

```python
documents = sent_tokenize(corpus)
```

```python
type(documents)
```

```python
for sentence in documents:
    print(sentence)
```

```python
## Tokenization
# Paragraph --> words
# sentence --> words

# word_tokenize -> Function used to break a sentence
#                  into individual words (called tokens)
from nltk.tokenize import word_tokenize 
```

```python
word_tokenize(corpus)
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

```python
# wordpunct_tokenize -> Function that splits text into words
#                       and punctuation separately

# It separates punctuation marks like ., ', !, ?, etc.
# from words and treats them as individual tokens.

from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus)
```

```python

from nltk.tokenize import TreebankWordTokenizer
```

```python
# TreebankWordTokenizer -> A tokenizer based on Penn Treebank rules
#                          used to split sentences into words.

# It handles punctuation and contractions more intelligently.

# Example:
# Sentence: "I can't do this."

# Output:
# ['I', 'ca', "n't", 'do', 'this', '.']

tokenizer = TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus)
```

### Difference Between `word_tokenize`, `wordpunct_tokenize`, and `TreebankWordTokenizer`

| Tokenizer | Description | Example Output for `"Hello, I'm learning NLP!"` |
|-----------|-------------|-----------------------------------------------|
| **word_tokenize** | Standard tokenizer in NLTK that splits text into words and punctuation using the Punkt model. | `['Hello', ',', 'I', "'m", 'learning', 'NLP', '!']` |
| **wordpunct_tokenize** | Splits words and punctuation strictly using punctuation rules. It separates punctuation completely. | `['Hello', ',', 'I', "'", 'm', 'learning', 'NLP', '!']` |
| **TreebankWordTokenizer** | Uses Penn Treebank rules and handles contractions more linguistically. | `['Hello', ',', 'I', "'m", 'learning', 'NLP', '!']` |
