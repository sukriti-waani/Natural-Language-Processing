## N-grams in NLP
**N-grams** are a text representation technique used in **Natural Language Processing (NLP)** where a sequence of **N consecutive words** from a sentence is used as a feature.

Here, **N** represents the number of words grouped together.

N-grams help capture **context and word order**, which methods like **Bag of Words cannot capture**.

---

## Example Sentence

```
I love learning NLP
```

---

## Types of N-grams

### 1. Unigram (1-gram)

A **Unigram** consists of **one word at a time**.

Example:

| Unigrams |
|---------|
| I |
| love |
| learning |
| NLP |

---

### 2. Bigram (2-gram)

A **Bigram** consists of **two consecutive words**.

Example:

| Bigrams |
|--------|
| I love |
| love learning |
| learning NLP |

---

### 3. Trigram (3-gram)

A **Trigram** consists of **three consecutive words**.

Example:

| Trigrams |
|---------|
| I love learning |
| love learning NLP |

---

### 4. N-gram (General Form)

An **N-gram** refers to any sequence of **N words together**.

Examples:

| N | Name | Example |
|---|------|--------|
| 1 | Unigram | NLP |
| 2 | Bigram | love NLP |
| 3 | Trigram | learning NLP techniques |
| 4 | Four-gram | I love learning NLP |

---

## Why N-grams are Useful

N-grams help capture **word order and context**.

Example:

```
New York
```

Bigram representation keeps the phrase together, while **Bag of Words would treat them separately**.

---

## Advantages

| Advantage | Explanation |
|----------|-------------|
| Captures context | Keeps neighboring words together |
| Improves language modeling | Helps predict next words |
| Useful in text classification | Provides more meaningful features |

---

## Disadvantages

| Disadvantage | Explanation |
|-------------|-------------|
| Large feature space | Vocabulary grows quickly |
| Sparse data problem | Many combinations may rarely appear |
| Higher computational cost | More memory and processing required |

---

## Example Comparison

Sentence:

```
Machine learning is powerful
```

| Method | Representation |
|------|---------------|
| Unigram | machine, learning, is, powerful |
| Bigram | machine learning, learning is, is powerful |
| Trigram | machine learning is, learning is powerful |

---

## Conclusion

N-grams are an important **text feature extraction technique in NLP** that capture **word sequences and context**, making them more informative than simple Bag of Words representations.
