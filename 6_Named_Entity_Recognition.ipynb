## Named Entity Recognition (NER)
**Named Entity Recognition (NER)** is a Natural Language Processing (NLP) technique used to **identify and classify important entities in text** into predefined categories.

These entities usually include:

- Person names
- Locations
- Organizations
- Dates
- Monetary values
- Percentages

NER helps computers understand **important real-world objects mentioned in a sentence**.

---

## Example

Sentence:

```
Barack Obama was born in Hawaii and served as the President of the United States.
```

NER Output:

| Word | Entity Type |
|-----|-------------|
| Barack Obama | Person |
| Hawaii | Location |
| United States | Location |

---

## Common Types of Named Entities

| Entity Type | Description | Example |
|-------------|-------------|---------|
| PERSON | Name of a person | Elon Musk |
| ORGANIZATION | Company or institution | Google |
| LOCATION | Place or geographical area | India |
| DATE | Date or time expression | 12 July 2024 |
| MONEY | Monetary value | $500 |
| PERCENT | Percentage | 45% |

---

## Why NER is Important

NER is widely used in many NLP applications:

- **Information extraction**
- **Search engines**
- **Chatbots**
- **Question answering systems**
- **News article analysis**

---

## Simple Example

Sentence:

```
Apple was founded by Steve Jobs in California.
```

NER identifies:

| Entity | Type |
|------|------|
| Apple | Organization |
| Steve Jobs | Person |
| California | Location |

---

## Conclusion

Named Entity Recognition helps computers **identify important entities in text and understand their meaning**, making it a key task in **Natural Language Processing**.

````python
sentence="The Eiffel Tower was built from 1887 to 1889 by Gustave Eiffel, whose company specialized in building metal frameworks and structures."
````

````python
import nltk
words = nltk.word_tokenize(sentence)
````

````python
tag_elements = nltk.pos_tag(words)
````

````python
import nltk

nltk.download('punkt')
# Required for sentence and word tokenization

nltk.download('averaged_perceptron_tagger')
# Required for POS tagging

nltk.download('maxent_ne_chunker')
# Named Entity Recognition model

nltk.download('maxent_ne_chunker_tab')
# Additional NER model tables required in new NLTK versions

nltk.download('words')
# Word dictionary used by the NER system
````

````python
# `ne_chunk()` is used in NLP to **detect and classify important named entities in text**, helping machines understand real-world information from sentences.

nltk.ne_chunk(tag_elements)
````
