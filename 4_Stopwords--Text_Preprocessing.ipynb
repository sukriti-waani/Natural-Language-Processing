## Stopwords in Text Preprocessing
**Stopwords** are very common words in a language that usually **do not carry important meaning** in a sentence.  
During **text preprocessing in Natural Language Processing (NLP)**, these words are often removed to focus on the important words.

Examples of stopwords:    
the, is, in, at, on, and, a, an

These words occur very frequently but **do not contribute much to understanding the main idea of the text**.

---

## Example

**Original Sentence**

```
The cat is sitting on the mat
```

**After Removing Stopwords**

```
cat sitting mat
```

Here, words like **the, is, on** are removed because they are stopwords.

---

## Why Stopwords Are Removed

Removing stopwords helps to:

- Reduce **text size**
- Improve **processing speed**
- Focus on **important keywords**
- Improve **machine learning model performance**

---

## Advantages

| Advantage | Explanation |
|----------|-------------|
| Reduces data size | Removes unnecessary frequent words |
| Faster processing | Less text for the model to analyze |
| Better focus | Keeps only meaningful words |

---

## Disadvantages

| Disadvantage | Explanation |
|-------------|-------------|
| May remove useful context | Some stopwords can be important in certain sentences |
| Not suitable for all tasks | Some NLP tasks require stopwords |

---

## Conclusion

Stopwords removal is an important **text preprocessing technique in NLP** that helps simplify text and focus on **meaningful words for analysis**.

````python
from nltk.stem import PorterStemmer
````

````python
from nltk.corpus import stopwords
````

````python
import nltk 
nltk.download('stopwords')
````

````python
from nltk.stem import PorterStemmer
````

````python
stemmer = PorterStemmer()
````

````python
sentences = nltk.sent_tokenize(paragraph)
# Splits the given paragraph into individual sentences
# Uses NLTK's sentence tokenizer based on punctuation and language rules
````

````python
type(sentences)
````

````python
# Apply Stopwords removal and then apply stemming

for i in range(len(sentences)):  
    # Loop through each sentence in the sentences list

    words = nltk.word_tokenize(sentences[i])  
    # Convert the sentence into individual words (tokenization)

    words = [stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]  
    # Remove stopwords from the sentence
    # Apply stemming to each remaining word
    # Create a new list of processed words

    sentences[i] = ' '.join(words)  
    # Join all processed words back into a sentence
    # Update the sentence in the sentences list
    
    
    
# Simple code
    
# Apply stopword removal and stemming

# for i in range(len(sentences)):
#     # Go through each sentence one by one
    
#     words = nltk.word_tokenize(sentences[i])
#     # Break the sentence into individual words
    
#     filtered_words = []
#     # Create an empty list to store processed words
    
#     for word in words:
#         # Check each word in the sentence
        
#         if word not in stopwords.words('english'):
#             # If the word is not a stopword
            
#             stem_word = stemmer.stem(word)
#             # Apply stemming to convert word to its root form
            
#             filtered_words.append(stem_word)
#             # Add the processed word to the list
    
#     sentences[i] = " ".join(filtered_words)
#     # Join the processed words back into a sentence
````
