```python
import pandas as pd 
messages = pd.read_csv('SpamClassifier-master/SMSSpamCollection.txt', sep='\t', names=["label", "message"])
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
import re
import nltk
nltk.download('stopwords')
```

```python
corpus = []
# Create an empty list to store the processed sentences

for i in range(0, len(messages)):
    # Loop through all rows of the dataset (messages)

    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    # Remove everything except alphabets using regular expression
    # Replace numbers, punctuation, symbols with space

    review = review.lower()
    # Convert all text to lowercase to maintain consistency

    review = review.split()
    # Split the sentence into individual words (tokenization)

    review = [lemmatizer.lemmatize(word) for word in review ]
    # Apply lemmatization  words to convert them to root form

    review = ' '.join(review)
    # Join all processed words back into a single sentence

    corpus.append(review)
    # Add the cleaned and processed sentence into the corpus list
```

```python
[[i, j, k] for i, j, k in zip(list(map(len, corpus)), corpus, messages['message']) if i < 1]
# Create a list using list comprehension

# map(len, corpus)
# Calculate the length of each processed document in corpus

# list(map(len, corpus))
# Convert the lengths into a list

# zip(...)
# Combine three items together:
# i → length of processed text
# j → processed text from corpus
# k → original message from dataset

# if i < 1
# Filter documents where the processed text length is less than 1
# (i.e., empty documents after preprocessing)

# Result:
# Returns rows showing:
# [length, processed_text, original_message]
# This helps identify messages that became empty after preprocessing
```

```python
from nltk import sent_tokenize
# It is used to split a paragraph into individual sentences

from gensim.utils import simple_preprocess
# It is used to clean and tokenize text into lowercase words
# Removes punctuation and special characters automatically
```

```python
words = []
# Create an empty list to store tokenized words

for sent in corpus:
    # Loop through each document/sentence in the corpus

    sent_token = sent_tokenize(sent)
    # Split the document into individual sentences

    for sent in sent_token:
        # Loop through each sentence obtained after tokenization

        words.append(simple_preprocess(sent))
        # simple_preprocess() cleans and converts the sentence into a list of lowercase words
        # It removes punctuation and special characters
        # The processed list of words is added to the "words" list
```

```python
# train Word2vec

model = gensim.models.Word2Vec(words)
# Create and train a Word2Vec model using the tokenized sentences stored in "words"

# gensim.models.Word2Vec
# Word2Vec algorithm used to learn vector representations (embeddings) of words

# words
# List of tokenized sentences where each sentence is a list of words

# The model learns semantic relationships between words
# Similar words will have similar vector representations
```

```python
import numpy as np
```

```python
def avg_word2vec(doc):
    # Function to compute the average Word2Vec vector for a document (list of words)

    vectors = [model.wv[word] for word in doc if word in model.wv.index_to_key]
    # Collect word vectors for all words in the document
    # Only include words that exist in the Word2Vec vocabulary

    if len(vectors) == 0:
        return np.zeros(model.vector_size)
        # If no words are found in the vocabulary, return a zero vector
        # model.vector_size gives the dimension of the word embeddings (e.g., 100)

    return np.mean(vectors, axis=0)
    # Compute the average of all word vectors along axis=0
    # This produces a single vector representing the entire document
```

```python
X = []
# Create empty list to store document vectors

for i in tqdm(range(len(words))):
# Loop through all tokenized sentences with progress bar

    print("Hello", i)
# Print the current index (optional)

    X.append(avg_word2vec(words[i]))
# Call the avg_word2vec function on the current sentence
# Append the resulting averaged word vector to list X
```

```python
# Dependent Features (Output Features)

y = messages[list(map(lambda x: len(x) > 0, corpus))]
# Select only those rows from the original dataset where the processed text in "corpus" is not empty
# map(lambda x: len(x) > 0, corpus) checks if each document in corpus has length greater than 0
# This avoids training the model on empty documents

y = pd.get_dummies(y['label'])
# Convert the 'label' column (ham/spam) into dummy variables (one-hot encoding)
# Example:
# ham  → [1,0]
# spam → [0,1]

y = y.iloc[:,0].values
# Select the first column from the dummy variables
# Convert it into a NumPy array for machine learning models
# Final result: y becomes a 1D array of labels
```
