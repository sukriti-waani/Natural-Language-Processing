## **`Parts of Speech (POS) Tagging`**
**Parts of Speech (POS) tagging** is a Natural Language Processing (NLP) technique used to **identify the grammatical role of each word in a sentence**.

It labels words based on their **function in the sentence**, such as noun, verb, adjective, etc.

---

## Example

Sentence:

```
The cat is playing in the garden
```

POS tagging result:

| Word | POS Tag | Meaning |
|-----|--------|--------|
| The | DT | Determiner |
| cat | NN | Noun |
| is | VBZ | Verb |
| playing | VBG | Verb (present participle) |
| in | IN | Preposition |
| the | DT | Determiner |
| garden | NN | Noun |

---

## Common POS Tags

| Tag | Full Form | Example |
|----|-----------|--------|
| NN | Noun | cat, book |
| NNS | Plural Noun | cars, dogs |
| VB | Verb (base form) | run, eat |
| VBD | Verb (past tense) | ran, ate |
| VBG | Verb (gerund) | running |
| JJ | Adjective | big, happy |
| RB | Adverb | quickly, slowly |
| PRP | Pronoun | he, she, they |
| DT | Determiner | the, a, an |
| IN | Preposition | in, on, at |

---

## Why POS Tagging is Important

POS tagging helps in:

- **Understanding sentence structure**
- **Text analysis**
- **Machine translation**
- **Chatbots and virtual assistants**
- **Information extraction**

---

## Simple Example

Sentence:

```
She runs quickly
```

POS Tags:

| Word | POS |
|-----|-----|
| She | PRP |
| runs | VBZ |
| quickly | RB |

---

## Conclusion

POS tagging helps computers **understand the grammatical structure of a sentence**, which is an important step in **Natural Language Processing tasks**.

````python
## Speech Of DR APJ Abdul Kalam
paragraph = """I have three visions for India. In 3000 years of our history, people from all over 
               the world have come and invaded us, captured our lands, conquered our minds. 
               From Alexander onwards, the Greeks, the Turks, the Moguls, the Portuguese, the British,
               the French, the Dutch, all of them came and looted us, took over what was ours. 
               Yet we have not done this to any other nation. We have not conquered anyone. 
               We have not grabbed their land, their culture, 
               their history and tried to enforce our way of life on them. 
               Why? Because we respect the freedom of others.That is why my 
               first vision is that of freedom. I believe that India got its first vision of 
               this in 1857, when we started the War of Independence. It is this freedom that
               we must protect and nurture and build on. If we are not free, no one will respect us.
               My second vision for India’s development. For fifty years we have been a developing nation.
               It is time we see ourselves as a developed nation. We are among the top 5 nations of the world
               in terms of GDP. We have a 10 percent growth rate in most areas. Our poverty levels are falling.
               Our achievements are being globally recognised today. Yet we lack the self-confidence to
               see ourselves as a developed nation, self-reliant and self-assured. Isn’t this incorrect?
               I have a third vision. India must stand up to the world. Because I believe that unless India 
               stands up to the world, no one will respect us. Only strength respects strength. We must be 
               strong not only as a military power but also as an economic power. Both must go hand-in-hand. 
               My good fortune was to have worked with three great minds. Dr. Vikram Sarabhai of the Dept. of 
               space, Professor Satish Dhawan, who succeeded him and Dr. Brahm Prakash, father of nuclear material.
               I was lucky to have worked with all three of them closely and consider this the great opportunity of my life. 
               I see four milestones in my career"""
````

````python
import nltk
from nltk.corpus import stopwords
sentences = nltk.sent_tokenize(paragraph)
````

````python
sentences
````

````python
import nltk
nltk.download('averaged_perceptron_tagger_eng')
# Downloads the pre-trained POS tagging model used by NLTK
# This model helps identify Parts of Speech (noun, verb, adjective, etc.) in a sentence
# Required when using nltk.pos_tag() for POS tagging
# Needs to be downloaded once and stored locally
````

````python
# We will find the POS (Parts of Speech) tags

stop_words = set(stopwords.words('english'))
# Load English stopwords once and convert them into a set for faster checking

for i in range(len(sentences)):
    # Loop through each sentence in the sentences list
    
    words = nltk.word_tokenize(sentences[i])
    # Break the sentence into individual words (tokenization)

    words = [word for word in words if word not in stop_words]
    # Remove stopwords like "the", "is", "and" from the list of words
    
    pos_tag = nltk.pos_tag(words)
    # Assign Parts of Speech (POS) tags to each word
    # Example: noun, verb, adjective, etc.

    print(pos_tag)
    # Print the word along with its POS tag
````

````python
sentence = "Taj Mahal is a beautiful Monument"

words = nltk.word_tokenize(sentence)
# Convert sentence into list of words

nltk.pos_tag_sents([words])
# pos_tag_sents expects a list of sentences (each sentence is a list of words)
````

````python
nltk.pos_tag_sents([words])
````

````python
sentence = "Taj Mahal is a beautiful Monument"

chars = list(sentence)
# Convert the sentence into individual characters

pos_tags = nltk.pos_tag(chars)
# Apply POS tagging to each character

print(pos_tags)
````
