## One-Hot Encoding in NLP
**One-Hot Encoding** is a technique used in Natural Language Processing (NLP) to convert **words into numerical vectors** so that machine learning models can understand text data.

In this method, each word in the vocabulary is represented by a **binary vector** where:

- One position is **1**
- All other positions are **0**

The length of the vector equals the **size of the vocabulary**.

---

## Example

Sentence:

```
I love NLP
```

Vocabulary:

| Word |
|-----|
| I |
| love |
| NLP |

One-Hot Representation:

| Word | Vector |
|-----|-------|
| I | [1, 0, 0] |
| love | [0, 1, 0] |
| NLP | [0, 0, 1] |

Here each word has its **own unique vector representation**.

---

## Intuition (Simple Idea)

Think of One-Hot Encoding like **assigning a unique ID position to each word**.

Example vocabulary:

| Word | Position |
|-----|---------|
| apple | 1 |
| banana | 2 |
| mango | 3 |

Vectors:

| Word | One-Hot Vector |
|-----|---------------|
| apple | [1,0,0] |
| banana | [0,1,0] |
| mango | [0,0,1] |

Each word activates **only one position in the vector**.

---

## Advantages

| Advantage | Explanation |
|----------|-------------|
| Simple to implement | Easy method to convert text into numbers |
| No assumptions about relationships | Treats each word independently |
| Works well for small vocabulary | Useful in simple NLP models |

---

## Disadvantages

| Disadvantage | Explanation |
|-------------|-------------|
| High dimensional vectors | Vector size equals vocabulary size |
| Sparse representation | Most values in the vector are 0 |
| No semantic meaning | Cannot capture similarity between words |
| Memory inefficient | Large vocabulary increases memory usage |

Example:

```
king → [0,0,1,0,0]
queen → [0,0,0,1,0]
```

Even though **king and queen are related**, their vectors do not reflect this relationship.

---

## Conclusion

One-Hot Encoding is a **basic word representation technique in NLP** that converts words into binary vectors.  
However, it does **not capture semantic relationships between words**, which is why more advanced methods like **Word Embeddings (Word2Vec, GloVe, BERT)** are often used.
