## **`Stemming`**
Stemming is the process of reducing a word to its root or base form by removing prefixes or suffixes.

````python
# Classification Problem
## Comments of product is a positive review or negative review
## Reviews----> eating, eat,eaten --> eat
words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]
````

#### **`Porter Stemmer`**
Porter Stemmer is a stemming algorithm used in NLP to reduce words to their root (stem) form by removing common suffixes.

````python
from nltk.stem import PorterStemmer
````

````python
stemming = PorterStemmer()
````

````python
# Loop through each word in the list called "words"
for word in words:
    # For every word in the list:
    # 1. word -> the original word
    # 2. stemming.stem(word) -> converts the word into its stem/root form
    print(word + " -----> " + stemming.stem(word))
````

````python
stemming.stem('congratulations')
````

````python
stemming.stem('sitting')
````

#### **`Disadvantages of Stemming`**

1. **Not always a real word** - The stemmed result may not be a meaningful word.   
Example:   
studies → studi

2. **Loss of meaning** - Different words with different meanings may become the same stem.   
Example:  
universe → univers 
university → univers

3. **Over-stemming** - The algorithm may remove too many characters.

4. L**ess accurate than lemmatization** - Stemming only removes suffixes without understanding grammar.

#### **`RegexpStemmer (Regular Expression Stemmer)`**

**RegexpStemmer** is a class in NLTK used for stemming words using **regular expressions (patterns)**.

It removes **prefixes or suffixes from words based on a pattern defined by the user**.

---

### Simple Idea

Instead of using a fixed algorithm like **Porter Stemmer**, `RegexpStemmer` removes parts of a word that match a **specific pattern**.

This means **you control which part of the word should be removed**.

---

### Example : 

Suppose we want to remove **"ing"** from words.

```python
from nltk.stem import RegexpStemmer

stemmer = RegexpStemmer('ing$')

print(stemmer.stem("playing"))

````python
from nltk.stem import RegexpStemmer
````

````python
# Creating a RegexpStemmer object

# 'ing$|s$|e$|able$' -> Regular expression pattern
# It removes these suffixes if they appear at the END of a word:
# ing  (playing → play)
# s    (cats → cat)
# e    (make → mak)
# able (readable → read)

# | -> means "OR" in regex (match any one of these patterns)
# $ -> means the pattern must appear at the end of the word
# min=4 -> stemming will only be applied if the word length is at least 4 characters

reg_stemmer = RegexpStemmer('ing$|s$|e$|able$', min=4)
````

````python
reg_stemmer.stem('eating')
````

````python
reg_stemmer.stem('ingeating')
````

#### **`Snowball Stemmer`**

Snowball Stemmer is a stemming algorithm used in Natural Language Processing (NLP) to reduce words to their root (stem) form by removing suffixes. It is an improved and more powerful version of the Porter Stemmer and supports multiple languages such as English, French, German, Spanish, etc.

The purpose of Snowball Stemmer is to reduce different forms of a word to a common base word so that machines can process text more efficiently.

For example, words like “playing”, “played”, and “plays” can all be reduced to the root word “play”.

````python
from nltk.stem import SnowballStemmer
````

````python
snowballstemmer = SnowballStemmer('english')
````

````python
for word in words:
    print(word + " ----> " + snowballstemmer.stem(word))
````

````python
stemming.stem("Fairly"),stemming.stem("sportingly")
````

````python
snowballstemmer.stem("Fairly"),snowballstemmer.stem("sportingly")
````

````python
snowballstemmer.stem('goes')
````

````python
snowballstemmer.stem('going')
````

````python
stemming.stem('goes')
````

# Key Differences Between the Stemmers

| Feature | Porter Stemmer | Regexp Stemmer | Snowball Stemmer |
|------|------|------|------|
| Approach | Rule-based algorithm | Pattern matching using regex | Improved rule-based algorithm |
| Developer | Martin Porter | NLTK | Martin Porter |
| Language Support | Mainly English | Depends on pattern | Multiple languages |
| Accuracy | Moderate | Depends on regex pattern | Better than Porter |
| Speed | Moderate | Fast | Faster than Porter |
| Complexity | Medium | Very simple | Medium |

---

# Disadvantages Comparison

| Stemmer | Disadvantages |
|------|------|
| Porter Stemmer | May produce incorrect stems (studies → studi), removes too many letters sometimes, limited mainly to English |
| Regexp Stemmer | Depends completely on user patterns, ignores linguistic rules, may generate incorrect stems |
| Snowball Stemmer | Still may produce non-dictionary stems, slightly more complex than simple stemmers |
