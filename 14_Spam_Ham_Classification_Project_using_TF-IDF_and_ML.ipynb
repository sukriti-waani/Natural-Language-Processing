```python
import pandas as pd 
messages = pd.read_csv('SpamClassifier-master/SMSSpamCollection.txt', sep='\t', names=["label", "message"])
```

```python
messages
```

```python
# Data cleaning and preprocessing

import re 
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
# Import the list of stopwords (common words like "the", "is", "and") from NLTK

from nltk.stem.porter import PorterStemmer
# Import PorterStemmer which is used to reduce words to their root form (stemming)

ps = PorterStemmer()
# Create an object of PorterStemmer to apply stemming on words
```

```python
corpus = []
# Create an empty list to store the processed sentences

for i in range(0, len(messages)):
    # Loop through all rows of the dataset (messages)

    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    # Remove everything except alphabets using regular expression
    # Replace numbers, punctuation, symbols with space

    review = review.lower()
    # Convert all text to lowercase to maintain consistency

    review = review.split()
    # Split the sentence into individual words (tokenization)

    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    # Remove stopwords like "the", "is", "and"
    # Apply stemming to remaining words to convert them to root form

    review = ' '.join(review)
    # Join all processed words back into a single sentence

    corpus.append(review)
    # Add the cleaned and processed sentence into the corpus list
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(corpus, y, test_size=0.20)
```

## **Creating the TF-IDF Model**

```python
# Train Test Split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(corpus, y, test_size=0.20)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tv = TfidfVectorizer(max_features=2500, ngram_range=(1, 2))

X_train = tv.fit_transform(X_train).toarray()
X_test = tv.transform(X_test).toarray()
```

```python
tv.vocabulary_
```

```python
from sklearn.naive_bayes import MultinomialNB

spam_tfidf_model = MultinomialNB().fit(X_train, y_train)
```

```python
# Prediction 

y_pred = spam_tfidf_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report

accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
