# Bag of Words (BoW) in NLP
**Bag of Words (BoW)** is a technique used in **Natural Language Processing (NLP)** to convert text into **numerical feature vectors**.

It represents a document by **counting how many times each word appears**, while **ignoring grammar and word order**.

The model only focuses on the **frequency of words in a document**.

---

# Intuition (Simple Idea)

Imagine we have several sentences and we want to teach a machine to understand them.

Instead of understanding grammar, the machine simply asks:

- Which words appear?
- How many times do they appear?

It then builds a **word frequency table**.

So the sentence is treated like a **bag of words**, where **order does not matter**.

Example:

```
Sentence 1: I love NLP
Sentence 2: I love machine learning
```

The machine just collects words like items in a bag.

---

# Steps of Bag of Words

Step 1: **Collect the Text Data**

Example sentences:

```
D1: I love NLP
D2: I love machine learning
```

---

Step 2: **Text Preprocessing**

- Convert to lowercase
- Remove punctuation
- Tokenization
- Remove stopwords (optional)

Example:

```
D1: love nlp
D2: love machine learning
```

---

Step 3: **Create Vocabulary**

Vocabulary = all unique words in all documents.

| Vocabulary |
|------------|
| love |
| nlp |
| machine |
| learning |

---

Step 4: **Create Word Count Vectors**

Now count the number of times each word appears in each document.

| Document | love | nlp | machine | learning |
|---------|------|-----|---------|----------|
| D1 | 1 | 1 | 0 | 0 |
| D2 | 1 | 0 | 1 | 1 |

These rows become the **numerical representation of documents**.

---

# Numerical Example (Step by Step)

Suppose we have three sentences:

```
S1: Cats like milk
S2: Dogs like milk
S3: Cats and dogs
```

---

### Step 1: Preprocess

```
cats like milk
dogs like milk
cats dogs
```

---

### Step 2: Vocabulary

| Words |
|------|
| cats |
| dogs |
| like |
| milk |

---

### Step 3: Create Word Count Table

| Sentence | cats | dogs | like | milk |
|---------|------|------|------|------|
| S1 | 1 | 0 | 1 | 1 |
| S2 | 0 | 1 | 1 | 1 |
| S3 | 1 | 1 | 0 | 0 |

This table is the **Bag of Words representation**.

---

# Key Characteristics

- Word order is ignored
- Only word frequency matters
- Each document becomes a **vector**

Example:

```
cats like milk
```

Vector representation:

```
[1, 0, 1, 1]
```

---

# Advantages

| Advantage | Explanation |
|-----------|-------------|
| Simple to understand | Easy method to convert text into numbers |
| Easy to implement | Works with basic machine learning models |
| Works well for small datasets | Useful for basic NLP tasks |
| Fast computation | Simple word counting approach |

---

# Disadvantages

| Disadvantage | Explanation |
|-------------|-------------|
| Ignores word order | Sentence meaning may change but vector remains same |
| High dimensional vectors | Large vocabulary leads to very large feature vectors |
| Sparse matrix problem | Most values are zero |
| No semantic meaning | Cannot understand similarity between words |

Example:

```
I love dogs
I love cats
```

Vectors:

| Sentence | dogs | cats |
|---------|------|------|
| I love dogs | 1 | 0 |
| I love cats | 0 | 1 |

Even though **cats and dogs are similar**, BoW treats them as completely unrelated.

---

# Example Showing Limitation

```
Sentence A: Dog bites man
Sentence B: Man bites dog
```

BoW representation for both:

```
[dog=1, bites=1, man=1]
```

Both sentences look **identical to the model**, even though the meaning is different.

---

# Conclusion

Bag of Words is a **basic text representation technique in NLP** that converts text into **word frequency vectors**.  

However, it has limitations such as:

- ignoring word order
- high dimensional vectors
- lack of semantic understanding

Because of these limitations, more advanced techniques like:

- **TF-IDF**
- **Word2Vec**
- **GloVe**
- **BERT**

are often used in modern NLP systems.

### Difference Between BoW and Binary BoW

| Feature | Bag of Words | Binary Bag of Words |
|-------|---------------|---------------------|
| Representation | Word frequency | Word presence |
| Values | Counts (0,1,2,...) | Only 0 or 1 |
| Information captured | Frequency of words | Only existence of words |

---

### Example

Sentence:

```
I love love NLP
```

Vocabulary:

| Words |
|------|
| I |
| love |
| NLP |

#### Bag of Words Representation

| Sentence | I | love | NLP |
|---------|---|------|-----|
| S1 | 1 | 2 | 1 |

Here **love appears twice**, so its value is **2**.

---

#### Binary Bag of Words Representation

| Sentence | I | love | NLP |
|---------|---|------|-----|
| S1 | 1 | 1 | 1 |

Here **love appears twice**, but Binary BoW only checks **whether the word exists**, so the value is **1**.

---

### Key Idea

- **BoW** → counts how many times a word appears  
- **Binary BoW** → checks only whether the word appears or not
