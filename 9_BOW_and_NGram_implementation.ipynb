```python
import pandas as pd 
messages = pd.read_csv('SpamClassifier-master/SMSSpamCollection.txt', sep='\t', names=["label", "message"])
```

```python
messages
```

```python
# Data cleaning and preprocessing

import re 
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
# Import the list of stopwords (common words like "the", "is", "and") from NLTK

from nltk.stem.porter import PorterStemmer
# Import PorterStemmer which is used to reduce words to their root form (stemming)

ps = PorterStemmer()
# Create an object of PorterStemmer to apply stemming on words
```

```python
corpus = []
# Create an empty list to store the processed sentences

for i in range(0, len(messages)):
    # Loop through all rows of the dataset (messages)

    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    # Remove everything except alphabets using regular expression
    # Replace numbers, punctuation, symbols with space

    review = review.lower()
    # Convert all text to lowercase to maintain consistency

    review = review.split()
    # Split the sentence into individual words (tokenization)

    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    # Remove stopwords like "the", "is", "and"
    # Apply stemming to remaining words to convert them to root form

    review = ' '.join(review)
    # Join all processed words back into a single sentence

    corpus.append(review)
    # Add the cleaned and processed sentence into the corpus list
```

```python
corpus
```

## **Create Bag Of Words**

```python
# Create the Bag of Words model

from sklearn.feature_extraction.text import CountVectorizer
# Import CountVectorizer from sklearn
# It is used to convert text data into numerical vectors using the Bag of Words technique


# for binary bow enable binary=True
cv = CountVectorizer(max_features=2500, binary=True)
# Create a CountVectorizer object
# max_features=2500 means the model will consider only the top 2500 most frequent words
# This helps reduce dimensionality and keeps the most important words in the vocabulary

# binary=True
# Uses Binary Bag of Words representation
# Instead of counting how many times a word appears, it only checks whether the word appears or not
# If the word appears → 1
# If the word does not appear → 0
```

```python
X = cv.fit_transform(corpus).toarray()
# fit_transform(corpus):
# 1) fit → learns the vocabulary (most frequent words up to max_features) from the corpus
# 2) transform → converts each sentence in the corpus into a Bag of Words vector

# toarray():
# Converts the sparse matrix output into a normal NumPy array

# X now contains the numerical Bag of Words representation of all sentences
# Rows → documents/sentences
# Columns → vocabulary words (features)
# Values → count of each word in the corresponding document
```
