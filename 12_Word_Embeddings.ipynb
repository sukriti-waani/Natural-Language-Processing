# Word Embeddings in NLP
**Word Embeddings** are numerical vector representations of words used in **Natural Language Processing (NLP)**.

Instead of representing words as simple counts (like Bag of Words), word embeddings represent words as **dense vectors that capture semantic meaning and relationships between words**.

Example:

```
king → [0.21, 0.45, 0.67, ...]
queen → [0.19, 0.44, 0.70, ...]
```

Words with similar meanings have **similar vector representations**.

Example:

```
king - man + woman ≈ queen
```

---

# Why Word Embeddings are Needed

Traditional methods like:

- One-Hot Encoding
- Bag of Words
- TF-IDF

have some problems:

| Problem | Explanation |
|-------|-------------|
| Sparse vectors | Most values are zero |
| Large dimensional vectors | Vocabulary size can be very large |
| No semantic meaning | Similar words are treated as unrelated |

Word embeddings solve these problems by learning **dense and meaningful vector representations**.

---

# Types of Word Embeddings

Word embeddings are mainly divided into two categories:

1. **Count-based / Frequency-based embeddings**
2. **Prediction-based embeddings (Deep Learning models)**

---

# 1. Count-Based / Frequency-Based Word Embeddings

These methods rely on **word co-occurrence statistics** in a corpus.

The idea is:

> Words appearing in similar contexts tend to have similar meanings.

Common count-based methods include:

- Bag of Words (BoW)
- TF-IDF
- Co-occurrence Matrix

---

## Co-occurrence Matrix Example

Sentence:

```
I like deep learning
I like NLP
```

Vocabulary:

```
I, like, deep, learning, NLP
```

Co-occurrence matrix:

| Word | I | like | deep | learning | NLP |
|----|----|----|----|----|----|
| I | 0 | 2 | 0 | 0 | 0 |
| like | 2 | 0 | 1 | 1 | 1 |
| deep | 0 | 1 | 0 | 1 | 0 |
| learning | 0 | 1 | 1 | 0 | 0 |
| NLP | 0 | 1 | 0 | 0 | 0 |

These vectors represent relationships between words.

However, count-based methods produce **very large sparse vectors**.

---

# 2. Prediction-Based Word Embeddings (Deep Learning)

Prediction-based embeddings use **neural networks to learn word representations**.

Instead of counting word occurrences, the model **learns vectors by predicting words in context**.

Popular models include:

- Word2Vec
- GloVe
- FastText

The most famous is **Word2Vec**, introduced by **Tomas Mikolov (Google, 2013)**.

---

# Word2Vec

**Word2Vec** is a neural network model used to learn word embeddings from large text corpora.

It converts each word into a **dense vector representation** that captures semantic meaning.

Word2Vec has two architectures:

1. **Continuous Bag of Words (CBOW)**
2. **Skip-gram**

---

# 1. Continuous Bag of Words (CBOW)

## Idea

CBOW predicts the **target word using its surrounding context words**.

Example sentence:

```
I love learning NLP
```

Suppose the target word is **learning**.

Context words:

```
I love ___ NLP
```

The model predicts the missing word **learning** using the surrounding words.

---

### CBOW Diagram

```
Context Words → Neural Network → Target Word
```

Example:

```
Input:  I, love, NLP
Output: learning
```

---

### Advantages of CBOW

- Faster training
- Works well with large datasets
- Good for frequent words

---

### Disadvantages of CBOW

- Poor performance for rare words
- Ignores some word order information

---

# 2. Skip-gram

## Idea

Skip-gram works **opposite of CBOW**.

Instead of predicting the center word, it **predicts surrounding context words from the target word**.

Example sentence:

```
I love learning NLP
```

Target word:

```
learning
```

The model predicts:

```
I, love, NLP
```

---

### Skip-gram Diagram

```
Target Word → Neural Network → Context Words
```

Example:

```
Input: learning
Output: I, love, NLP
```

---

### Advantages of Skip-gram

- Works well for **rare words**
- Captures better semantic relationships

---

### Disadvantages of Skip-gram

- Slower training
- Requires more computational resources

---

# Difference Between CBOW and Skip-gram

| Feature | CBOW | Skip-gram |
|------|------|------|
| Prediction | Context → Target | Target → Context |
| Training speed | Faster | Slower |
| Rare words | Poor performance | Better performance |
| Accuracy | Slightly lower | Higher |

---

# Example of Word Embedding Relationships

Word embeddings capture **semantic relationships**:

```
king - man + woman ≈ queen
```

Other examples:

```
Paris - France + Italy ≈ Rome
```

This shows embeddings understand **semantic similarity and relationships between words**.

---

# Advantages of Word Embeddings

| Advantage | Explanation |
|------|-------------|
| Dense vectors | Smaller representation compared to BoW |
| Captures semantic meaning | Similar words have similar vectors |
| Efficient for ML models | Works well in deep learning |
| Handles large vocabulary | Better representation than one-hot encoding |

---

# Disadvantages of Word Embeddings

| Disadvantage | Explanation |
|------|-------------|
| Requires large training data | Needs large corpus to learn good embeddings |
| Computationally expensive | Training neural networks requires resources |
| Context limitation | Basic embeddings do not consider sentence context fully |

---

# Applications of Word Embeddings

Word embeddings are used in:

- Machine translation
- Chatbots
- Question answering systems
- Sentiment analysis
- Text classification
- Search engines

---

# Conclusion

Word embeddings are an advanced way of representing words in NLP.

They improve over traditional methods like:

- One-Hot Encoding
- Bag of Words
- TF-IDF

by capturing **semantic meaning and relationships between words**, making them very useful in **modern NLP and deep learning models**.
